```python
import streamlit as st
import pickle
import pandas as pd
import urllib.parse

# ------------------ Functions ------------------

def fetch_poster(movie_name):
    try:
        # URL-encode the movie name for safe query string in the Google search
        movie_query = urllib.parse.quote_plus(movie_name + " movie poster")
        search_url = f"https://www.google.com/search?hl=en&tbm=isch&q={movie_query}"

        # Google will redirect to a page showing the image results, get the first result's URL
        return search_url

    except Exception as e:
        st.error(f"Error fetching poster for {movie_name}: {e}")
        return "https://via.placeholder.com/500x750?text=No+Image"  # Fallback

def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    recommended_movies = []
    recommended_movie_posters = []
    for i in movies_list:
        movie_name = movies.iloc[i[0]].title
        recommended_movies.append(movie_name)
        poster_url = fetch_poster(movie_name)
        recommended_movie_posters.append(poster_url)

    return recommended_movies, recommended_movie_posters

# ------------------ Main App ------------------

# Load data
movies_dict = pickle.load(open('movie_dict.pkl', 'rb'))
movies = pd.DataFrame(movies_dict)
similarity = pickle.load(open('similarity.pkl', 'rb'))

# App UI
st.set_page_config(page_title="Movie Recommender", page_icon="üé¨", layout="wide")
st.markdown("<h1 style='text-align: center; color: #E50914;'>üé¨ Movie Recommender System (Google Posters)</h1>", unsafe_allow_html=True)
st.markdown("### Pick a movie you like üëá", unsafe_allow_html=True)

movie_list = movies['title'].values
selected_movie_name = st.selectbox('', movie_list)

if st.button('üéØ Recommend'):
    with st.spinner('Fetching best matches for you... üé•'):
        recommended_movie_names, recommended_movie_posters = recommend(selected_movie_name)

    st.markdown("---")
    st.markdown("<h3 style='text-align: center;'>Movies You Might Love ‚ù§Ô∏è</h3>", unsafe_allow_html=True)

    cols = st.columns(5)
    for j in range(5):
        with cols[j]:
            # Use the search URL directly in the markdown for the clickable image
            movie_search_url = f"https://www.google.com/search?q={recommended_movie_names[j]}+movie+poster&tbm=isch"
            st.markdown(
                f'''
                <a href="{movie_search_url}" target="_blank" style="text-decoration: none;">
                    <img src="https://via.placeholder.com/500x750?text=Image+Not+Found" 
                         style="width: 220px; height: 330px; border-radius: 10px; 
                         transition: transform 0.2s;" 
                         onmouseover="this.style.transform='scale(1.05)';" 
                         onmouseout="this.style.transform='scale(1)';">
                </a>
                ''',
                unsafe_allow_html=True
            )
            st.caption(recommended_movie_names[j])
```
